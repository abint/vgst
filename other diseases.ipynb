```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as py
```

```python
data=pd.read_csv('data.csv')
data.dropna()

data.head()
```

```python
req=['Disease','Count']
data_d=data[req]
data_d=pd.get_dummies(data_d)
data_d.head()
```

```python
data_d.plot.pie(subplots=True)
py.show()
```
