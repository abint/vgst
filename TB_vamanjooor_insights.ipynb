```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("vgst/vamanjur/TBHospitalVamajurdata.csv")
agedata=data[["Age","Gender"]]
agedata['Gender']=agedata['Gender'].replace(['F','M'],['f','m'])
agedata

```

```python
import matplotlib.pyplot as plt
```

```python
counts = agedata.groupby('Age')['Gender'].count()
countage=agedata.groupby('Age').count()
countage.plot.pie(subplots=True)
plt.show()
```

```python
counts.plot.bar(subplots=True,figsize=(10,5))
plt.legend()
plt.show()
```

```python
gen=agedata.groupby('Gender').count()
gen.plot.pie(subplots=True)
plt.show()
```
